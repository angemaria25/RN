```python
import os
import shutil

# Ruta de la carpeta donde están tus imágenes mezcladas
input_folder = r"D:\Documentos\3RO\1ER Semestre\RN\datasets\Facade CMP\CMP_facade_DB_base\completly"

# Carpetas de salida
jpg_folder = os.path.join(input_folder, "images")
png_folder = os.path.join(input_folder, "masks")

# Crear carpetas si no existen
os.makedirs(jpg_folder, exist_ok=True)
os.makedirs(png_folder, exist_ok=True)

# Recorrer archivos
for filename in os.listdir(input_folder):
    file_path = os.path.join(input_folder, filename)

    # Ignorar carpetas
    if os.path.isdir(file_path):
        continue

    # Clasificar por extensión
    if filename.lower().endswith(".jpg"):
        shutil.move(file_path, os.path.join(jpg_folder, filename))
    elif filename.lower().endswith(".png"):
        shutil.move(file_path, os.path.join(png_folder, filename))

print("Clasificación completada.")
```
